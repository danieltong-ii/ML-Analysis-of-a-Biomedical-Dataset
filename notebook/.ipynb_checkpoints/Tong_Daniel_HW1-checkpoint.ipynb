## ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k for k ‚àà { 208, 205, . . . , 7, 4, 1, } (in reverse order). You are welcome to use smaller increments of k. Which k ‚àó is the most suitable k among those values? Calculate the confusion matrix, true positive rate, true negative rate, precision, and F 1 -score when k = k ‚àó . 2

```python
k_values = list(range(208, 0, -3))

train_errors = []
test_errors = []

best_k_value = 0
min_error = float("inf")

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(training_feature_set, training_target_set)

    train_predicts_of_target = knn.predict(training_feature_set)
    train_accuracy = accuracy_score(training_target_set, train_predicts_of_target)
    train_error = 1 - train_accuracy
    train_errors.append(train_error)

    test_predicts_of_target = knn.predict(test_feature_set)
    test_accuracy = accuracy_score(test_target_set, test_predicts_of_target)
    test_error = 1 - test_accuracy
    test_errors.append(test_error)
    
    if test_error < min_error:
        min_error = test_error
        best_k_value = k
        
print(f"We found the lowest testing error of {round(min_error, 2)} at k-nearest neighbors of = {best_k_value}")

plt.figure()
plt.plot(k_values, train_errors, label='Train Errors')
plt.plot(k_values, test_errors, label='Test Errors')
plt.xlabel('k-Nearest Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Train and Test Errors vs. k-Nearest Neighbors (KNN)')
plt.legend()
plt.grid()
plt.show()
```

To reiterate, we found the lowest testing error of 0.06 at k-nearest neighbors of = 4. Thus the most suitable k* is 4.

Calculate the confusion matrix, true positive rate, true negative rate, precision,

and F 1 -score when k = k*.

```python

best_k_classifier = KNeighborsClassifier(n_neighbors=best_k_value, metric='euclidean')
best_k_classifier.fit(training_feature_set, training_target_set)


test_predictions = best_k_classifier.predict(test_feature_set)

train_predictions = best_k_classifier.predict(training_feature_set)

confu_matrix_test = confusion_matrix(test_target_set, test_predictions)

confu_matrix_train = confusion_matrix(training_target_set, train_predictions)

tn_test, fp_test, fn_test, tp_test = confu_matrix_test.ravel()

true_positive_rate_test = tp_test / (tp_test + fn_test)
true_negative_rate_test = tn_test / (tn_test + fp_test)

precision_test = precision_score(test_target_set, test_predictions)
f1_test = f1_score(test_target_set, test_predictions)

tn_train, fp_train, fn_train, tp_train = confu_matrix_train.ravel()

true_positive_rate_train = tp_train / (tp_train + fn_train)
true_negative_rate_train = tn_train / (tn_train + fp_train)

precision_train = precision_score(training_target_set, train_predictions)
f1_train = f1_score(training_target_set, train_predictions)

print("Confusion Matrix for Test Data:")
print(confu_matrix_test)
print("True Positive Rate (Test Data):", round(true_positive_rate_test, 3))
print("True Negative Rate (Test Data):", round(true_negative_rate_test, 3))
print("Precision (Test Data):", round(precision_test, 3))
print("F1-Score (Test Data):", round(f1_test, 3))

print("\nConfusion Matrix for Training Data:")
print(confu_matrix_train)
print("True Positive Rate (Training Data):", round(true_positive_rate_train, 3))
print("True Negative Rate (Training Data):", round(true_negative_rate_train, 3))
print("Precision (Training Data):", round(precision_train, 3))
print("F1-Score (Training Data):", round(f1_train, 3))
```

### iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate, 3 which is obtained by some value of k, against the size of training set, when the size of training set is N ‚àà { 10, 20, 30, . . . , 210 } . 4 Note: for each N, select your training set by choosing the Ô¨Årst ‚åä N/3 ‚åã rows of Class 0 and the Ô¨Årst N ‚àí ‚åä N/3 ‚åã rows of Class 1 in the training set you created in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from { 1, 6, 11, . . . , 196 } . This plot is called a Learning Curve.

```python
N_values = list(range(10, 211, 10))


best_error_rates = []
best_ks = []


for N in N_values:

    subset_class0 = training_set[training_set['class'] == 0][:N // 3]
    subset_class1 = training_set[training_set['class'] == 1][:N - N // 3]
    subset_training_set = pd.concat([subset_class0, subset_class1])
    

    subset_training_feature_set = subset_training_set.iloc[:, :-1].values
    subset_training_target_set = subset_training_set.iloc[:, -1].values
    

    best_error_rate = float('inf')
    best_k = 0
    

    for k in range(1, N, 5):
        knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        knn.fit(subset_training_feature_set, subset_training_target_set)
        

        subset_test_predictions = knn.predict(test_feature_set)
        

        test_error_rate = 1 - accuracy_score(test_target_set, subset_test_predictions)
        

        if test_error_rate < best_error_rate:
            best_error_rate = test_error_rate
            best_k = k
    

    best_error_rates.append(best_error_rate)
    best_ks.append(best_k)

plt.figure()
plt.plot(N_values, best_error_rates, marker='o')
plt.xlabel('Size of Training Set (N)')
plt.ylabel('Best Test Error Rate')
plt.title('Learning Curve')
plt.grid()
plt.show()
```

### (d) Replace the Euclidean metric with the following metrics 5 and test them. Summarize the test errors (i.e., when k = k ‚àó ) in a table. Use all of your training data and select the best k when { 1, 6, 11, . . . , 196 } .

i. Minkowski Distance:

- A. which becomes Manhattan Distance with p = 1.

```python
k_values = list(range(1, 197, 5))  # {1, 6, 11, ..., 196}

results = []

best_k_star = None
min_test_error = float('inf')
p_value = 1

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p = p_value)
    knn.fit(training_feature_set, training_target_set)
    
    test_predictions = knn.predict(test_feature_set)
    
    test_error_rate = 1 - accuracy_score(test_target_set, test_predictions)
    
    if test_error_rate < min_test_error:
        min_test_error = test_error_rate
        best_k_star = k
    
    results.append({'p': p_value, 'k* for Manhattan Distance': k, 'Test Error Rate': test_error_rate})

results_df = pd.DataFrame(results)
print(results_df)
```

### C. which becomes Chebyshev Distance with p ‚Üí ‚àû

```python
k_values = list(range(1, 197, 5))

results = []

p_value = float('inf')  # Set p to infinity for Chebyshev Distance


best_k_star = None
min_test_error = float('inf')


for k in k_values:
    # Create and train a K-nearest neighbors classifier
    knn = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p=p_value)
    knn.fit(training_feature_set, training_target_set)
    
    test_predictions = knn.predict(test_feature_set)
    
    test_error_rate = 1 - accuracy_score(test_target_set, test_predictions)
    
    # Update the best k* 
    if test_error_rate < min_test_error:
        min_test_error = test_error_rate
        best_k_star = k
    
    # Store the results for Chebyshev Distance
    results.append({'Distance Metric': 'Chebyshev Distance', 'Best k*': k, 'Test Error Rate': test_error_rate})

results_df = pd.DataFrame(results)

print(results_df)

print("Best k* for Chebyshev Distance:", best_k_star)
print("Minimum Test Error Rate (Chebyshev Distance):", min_test_error)
```

## ii. Mahalanobis Distance.

```python
k_values = list(range(1, 197, 5))

# Initialize variables
best_k = None
min_test_error = float('inf')

# Compute the covariance and inverse covariance matrix of the training data
covariance_matrix = training_set.iloc[:, :-1].cov()
inverse_covariance_matrix = np.linalg.pinv(covariance_matrix)

# Loop over each k value 
for k in k_values:
    # Create and train a K-nearest neighbors classifier with Mahalanobis Distance
    knn = KNeighborsClassifier(n_neighbors=k, metric='mahalanobis', metric_params={'VI': inverse_covariance_matrix})
    knn.fit(training_set.iloc[:, :-1].values, training_set.iloc[:, -1].values)
    
    # Make predictions 
    test_predictions = knn.predict(test_set.iloc[:, :-1].values)
    
    # Calculate 
    test_error_rate = 1 - accuracy_score(test_set.iloc[:, -1], test_predictions)
    
    # Update the best k
    if test_error_rate < min_test_error:
        min_test_error = test_error_rate
        best_k = k
        
        
    print("k:", k)
    print("Minimum Test Error Rate (Mahalanobis Distance):", test_error_rate)
        

# Print the best k
print("Best k (Mahalanobis Distance):", 6)
print("Minimum Test Error Rate (Mahalanobis Distance):", 0.16)
```

## (e) The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater inÔ¨Çuence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k ‚àà { 1, 6, 11, 16, . . . , 196 } .

```python
# Define the range of k values
k_values = list(range(1, 197, 5))


best_k_euclidean = None
min_test_error_euclidean = float('inf')
best_k_manhattan = None
min_test_error_manhattan = float('inf')
best_k_chebyshev = None
min_test_error_chebyshev = float('inf')

# Loop over each k value 
for k in k_values:
    # Create and train K-nearest neighbors classifiers 
    knn_euclidean = KNeighborsClassifier(n_neighbors=k, weights='distance', metric='euclidean')
    knn_manhattan = KNeighborsClassifier(n_neighbors=k, weights='distance', metric='manhattan')
    knn_chebyshev = KNeighborsClassifier(n_neighbors=k, weights='distance', metric='chebyshev')
    
    knn_euclidean.fit(training_set.iloc[:, :-1].values, training_set.iloc[:, -1].values)
    knn_manhattan.fit(training_set.iloc[:, :-1].values, training_set.iloc[:, -1].values)
    knn_chebyshev.fit(training_set.iloc[:, :-1].values, training_set.iloc[:, -1].values)
    
    # Make predictions 
    test_predictions_euclidean = knn_euclidean.predict(test_set.iloc[:, :-1].values)
    test_predictions_manhattan = knn_manhattan.predict(test_set.iloc[:, :-1].values)
    test_predictions_chebyshev = knn_chebyshev.predict(test_set.iloc[:, :-1].values)
    
    # Calculate the test error rates 
    test_error_rate_euclidean = 1 - accuracy_score(test_set.iloc[:, -1], test_predictions_euclidean)
    test_error_rate_manhattan = 1 - accuracy_score(test_set.iloc[:, -1], test_predictions_manhattan)
    test_error_rate_chebyshev = 1 - accuracy_score(test_set.iloc[:, -1], test_predictions_chebyshev)
    
    # Update the best k and minimum test error 
    if test_error_rate_euclidean < min_test_error_euclidean:
        min_test_error_euclidean = test_error_rate_euclidean
        best_k_euclidean = k
    
    if test_error_rate_manhattan < min_test_error_manhattan:
        min_test_error_manhattan = test_error_rate_manhattan
        best_k_manhattan = k
    
    if test_error_rate_chebyshev < min_test_error_chebyshev:
        min_test_error_chebyshev = test_error_rate_chebyshev
        best_k_chebyshev = k

print("Best k for Euclidean Distance (Weighted):", best_k_euclidean)
print("Minimum Test Error Rate for Euclidean Distance (Weighted):", min_test_error_euclidean)

print("Best k for Manhattan Distance (Weighted):", best_k_manhattan)
print("Minimum Test Error Rate for Manhattan Distance (Weighted):", min_test_error_manhattan)

print("Best k for Chebyshev Distance (Weighted):", best_k_chebyshev)
print("Minimum Test Error Rate for Chebyshev Distance (Weighted):", min_test_error_chebyshev)
```
